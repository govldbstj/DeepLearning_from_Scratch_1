### 퍼셉트론으로 논리회로 만들기
AND, OR, NAND 게이트는 퍼셉트론으로 구현 가능.</br>
퍼셉트론의 구조는 AND, OR, NAND 게이트 모두에서 똑같다. 다른 건 가중치와 임계값.</br>
image.png
XOR은 불가능. 다층 퍼셉트론(비선형)으로 구현 가능.


###퍼셉트론 구현하기


```python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
```

```python
AND(0,0)
AND(1,0)
AND(0,1)
AND(1,1)
```

####가중합 < 임계값 = 가중합 - 편향 < 0
image.png

```python
import numpy as np

x = np.array([0,1])
w = np.array([0.5, 0.5])
b = -0.7
np.sum(w*x)
np.sum(w*x)+b
```

```python
def AND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(w*x)+b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, 0.5])
  b = 0.7
  tmp = np.sum(w*x)+b
  if tmp<=0:
    return 0
  else:
    return 1
```

```python
def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  tmp = np.sum(w*x)+b
  if tmp<=0:
    return 0
  else:
    return 1
```

####XOR게이트는 어떻게 구현할까?
퍼셉트론은 직선만 가능한데..

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```
