##**클래스 생성</br>**
(1) 생성자의 매개변수는 self 뒤에 붙인다.</br>
(2) 생성자에 속한 변수들을 사용할 땐 self를 사용해야 한다.</br>
(3) 클래스를 사용하기 위해서는 인스턴스를 만든다. 이때 필요에 따라 매개변수를 지정한다.</br>

```python
class Man:
  def __init__(self, name):
    self.name = name
    print("initialized!")

  def hello(self):
    print("Hello "+ self.name + "!")

  def goodbye(self):
    print("Good-bye "+ self.name + "!")

m = Man("David")
m.hello()
m.goodbye()
```

##**Numpy**

```python
import numpy as np
```

```python
# x와 y의 원소 수는 같아야 한다.
# 원소별 곱셈을 element-wise procuct라고 한다.

x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])

print(x+y)
print(x-y)
print(x*y)
print(x/y)
```

```python
# broadcast
# 넘파이 배열과 수치 하나(스칼라값)의 연산도 가능하다.

x = np.array([1.0, 2.0, 3.0])
print(x/2.0)
```

```python
# 넘파이로 다차원 행렬 만들기
A = np.array([[1,2], [3,4]])
print(A)
print(A.shape) # shape : 형상 (각 차원의 크기(원소 수))
print(A.dtype)
```

```python
B = np.array([[3,0], [0,6]])
print(A+B)
print(A*B)
```

```python
print(A)
print(A*10)
```

### 원소 접근
원소의 인덱스는 0부터 시작한다.

```python
X = np.array([[51,55], [14,19], [0,4]])
print(X)
print(X[0])
print(X[0][1])
```

```python
for row in X:
  print(row)

for i in X:
  print(i)

for element in X:
  print(element)
```

### 행렬에서 원하는 조건의 원소만 얻는 방법
1. 평탄화 flatten
2. 조건 걸어 찾기



```python
print(X)

X = X.flatten()
print(X)

X[np.array([0,2,4])] # 인덱스가 0, 2, 4인 원소를 찾아 프린트하기
```

```python
print(X>15)
print(X[X>15])
```

##**Matplotlib**

### pyplot

```python
import numpy as np
# plyplot : matplotlib의 내장함수, 간단한 2d 그래프 만드는 모듈
import matplotlib.pyplot as plt 

# np.arrange : python range랑 같은 역할. (for i in range)
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성 -> x좌표를 찍는다
y = np.sin(x) # sin함수를 통과시켜 y좌표를 찍는다

# 그래프 그리기
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = "sin", color = "pink")
plt.plot(x, y2, linestyle = "--", label = "cos", color = "purple")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend() ## 어떤 선이 어떤 함수인지 라벨 표시 (좌측 하단)
plt.show()
```
